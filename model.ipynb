```python
# dividing the dataset into dependent and independent variables

x=df.drop('Diagnosis',axis=1)
y=df['Diagnosis']
```

```python
pip install scikit-learn
```

```python
# dividing the dataset into train_test_split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
# appling the models
from sklearn.ensemble import RandomForestClassifier
# Choose and train a classification model (example: Random Forest)
rfc_model = RandomForestClassifier(n_estimators=100, random_state=42)  # Hyperparameters can be tuned
rfc_model.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = rfc_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
# Calculate model accuracy
model_accuracy = accuracy_score(y_test, y_pred)
print("The model accuracy:", model_accuracy)
```

```python
from sklearn.metrics import confusion_matrix

# Compute the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Print the confusion matrix
print("Confusion Matrix:")
print(conf_matrix)
```
